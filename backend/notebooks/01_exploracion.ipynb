# 01 - ExploraciÃ³n de datos: LaLiga 2025/26
Carga de datos desde FBref usando soccerdata.

```python
import soccerdata as sd
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 120)
```

## 1. Calendario de partidos â€” LaLiga 2025/26

```python
fbref = sd.FBref(leagues='ESP-La Liga', seasons='2526')

schedule = fbref.read_schedule()
print(f'Partidos cargados: {len(schedule)}')
print(f'Columnas: {list(schedule.columns)}')
schedule.head(10)
```

## 2. EstadÃ­sticas estÃ¡ndar de jugadores â€” LaLiga 2025/26

```python
player_stats = fbref.read_player_season_stats(stat_type='standard')
print(f'Jugadores cargados: {len(player_stats)}')
print(f'Columnas: {list(player_stats.columns)}')
player_stats.head(10)
```

## 3. InspecciÃ³n rÃ¡pida

```python
print('=== CALENDARIO ===')
print(schedule.dtypes)
print()
print('=== JUGADORES ===')
print(player_stats.dtypes)
```

```python
# Top 10 jugadores por goles
# Tras reset_index() todas las columnas son tuplas: ('player',''), ('team',''), ('Performance','Gls')
df = player_stats.reset_index()

top_goleadores = (
    df
    .sort_values(('Performance', 'Gls'), ascending=False)
    [[('player', ''), ('team', ''), ('Performance', 'Gls')]]
    .head(10)
)
top_goleadores.columns = ['jugador', 'equipo', 'goles']
print(top_goleadores.to_string(index=False))
```
